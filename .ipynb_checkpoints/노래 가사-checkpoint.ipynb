```python
driver = webdriver.Chrome()
driver.get('https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq=506712136')
driver.implicitly_wait(2)
```

```python
path = '#frm > div > table > tbody > tr'
num = len(driver.find_elements_by_css_selector(path))
```

```python
for ind in range(1,1+num): 
    driver.find_element_by_css_selector(f'#frm > div > table > tbody > tr:nth-child({ind}) > td:nth-child(4) > div > a').click()
    scroll_button = '#lyricArea > button'

#     print(driver.find_elements_by_css_selector(scroll_button))
    if driver.find_elements_by_css_selector(scroll_button):
        driver.find_elements_by_css_selector(scroll_button)[0].click()
        print('click')
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        tell = soup.select('#d_video_summary')[0].text
        print(re.compile('[ㄱ-ㅎ가-힣]+').findall(tell))
        print('='*150)
    else:
        pass
    driver.back()
```
