```python
def song_num(url):
    url = input()
    driver = webdriver.Chrome()
    driver.get(url)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    num = soup.select('#conts > div.section_contin > div.page_header > h5 > span')[0].text
    num = num.replace('(','')
    num = num.replace(')','')
    num = int(num)
    print(num)
```
