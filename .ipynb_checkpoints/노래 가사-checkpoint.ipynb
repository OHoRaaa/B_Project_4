```python
import requests
import requests as req
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
import time
```

```python
def lyrics(url):
    url = input()
    driver = webdriver.Chrome()
    driver.get(url)
    
    while True:
        for i in range(1,51):
            path = f'//*[@id="frm"]/div/table/tbody/tr[{i}]/td[4]/div/a'
            button = '//*[@id="lyricArea"]/button'

            driver.find_element_by_xpath(path).click()
            time.sleep(1)
            driver.find_element_by_xpath(button).click()
            time.sleep(1)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            tell = soup.select('#d_video_summary')[0].text
            print(re.compile('[ㄱ-ㅎ가-힣]+').findall(tell))
            print('='*150)
            driver.back()
```

```python
lyrics('url')
```
