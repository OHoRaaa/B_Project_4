```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from konlpy.tag import Kkma, Komoran, Okt, Hannanum
```

```python
okt = Okt()
```

```python
url = 'https://novel.naver.com/webnovel/detail?novelId=920930&volumeNo=159'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
lis = soup.select('div.detail_view_content')[0].text
liss = re.compile('[ㄱ-ㅎ가-힣]+').findall(lis)
print(liss)
print('='*120)
print('='*120)
print('='*120)
```

```python
liss = [j for j in liss if len(j) >= 2]

token=[]

for i in liss:
    token.append(okt.morphs(i))
    
```

```python
tokenize = []

for tok in token:
    for to in tok:
        if len(to) >=2:
            tokenize.append(to)
tokenize
```

```python
df = pd.read_excel('감성사전_편집.xlsx', index_col='Unnamed: 0')
df
```

```python
for i in tokenize:
    for j in df.index:
        if i == j:
            print(i)
```

```python
counter={}
for i in tokenize:
    for j in df.index:
        if i == j:
            if i in counter:
                counter[i] += 1
            else:
                counter[i] = 1
```

```python
emotion_score = {}

for key in counter.keys():
    for word in df.index:
        if key == word:
            for label in df.columns[df.loc[key,:] == 1]:
                if label in emotion_score:
                    emotion_score[label] += 1
                else:
                    emotion_score[label] = 1
```
